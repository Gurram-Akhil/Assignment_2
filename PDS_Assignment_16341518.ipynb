```python
#importing all the required libraries
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from datetime import datetime
```

```python
#Reading the dataset
df=pd.read_csv("train.csv")
df
```

```python
df=df.drop('Unnamed: 0', axis=1)
df.head()
```

Checking all null values in the data 

```python
df.isna().sum()
```

```python
#dropping New_Price as it has more null values
df = df.drop('New_Price', axis=1)
```

```python
df.head()
```

(b) Removing the characters from Mileage, Engine, Power and New_Price and converted them into float

```python
df['Mileage']=df['Mileage'].str.split(' ').str[0]
df['Engine']=df['Engine'].str.split(' ').str[0]
df['Power']=df['Power'].str.split(' ').str[0]
df['Seats']=df['Seats'].fillna(0)
```

```python
df['Mileage'] = df['Mileage']. astype(float)
df['Engine'] = df['Engine']. astype(float)
df['Power'] = df['Power']. astype(float)
df['Seats'] = df['Seats'] . astype(float)
```

```python
#replacing the Null values with mean
numeric_columns = ['Mileage', 'Engine', 'Power', 'Price','Seats']
imputer = SimpleImputer(strategy='mean')
df[numeric_columns] = imputer.fit_transform(df[numeric_columns])
```

```python
df
```

```python
#checking if the null values are replaced or not
df.isnull().sum()
```

C) Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot encoded value

```python
df = pd.get_dummies(df, columns=['Fuel_Type', 'Transmission'], prefix=['Fuel','Transmission'])
```

```python
df.head()
```

```python
col = ['Fuel_Diesel','Fuel_Electric','Fuel_Petrol','Transmission_Automatic','Transmission_Manual']
#converting spercified columns to integers
df[col] = df[col].astype(int)
```

```python
df
```

D) Create one more feature and add this column to the dataset (you can use mutate function in R for this). For example, you can calculate the current age of the car by subtracting “Year” value from the current year

```python
current_year = datetime.now().year
df['Current_Age'] = current_year - df['Year']
```
